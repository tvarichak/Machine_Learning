#### <span style='color:Black'> **Introduction** </span>

- Decision Trees were first proposed by Leo Breiman, a statistician at the University of California, Berkeley.
- Decision Trees are the foundation for many classical machine learning algorithms like Random Forests.
- Decision trees are now widely used in many applications for predictive modeling, including both **classification** and **regression**.



#### <span style='color:Black'> **Decision Tree algorithm terminology** </span>

<p align="center">
    <img src='./imgs/pic1.png' width = '50%'>
</p>

**Root Node:** It represents the entire population or sample. This further gets divided into two or more homogeneous sets.
Splitting. It is a process of dividing a node into two or more sub-nodes.
<br>

**Decision Node:** When a sub-node splits into further sub-nodes, then it is called a decision node.
<br>

**Leaf/Terminal Node:** Nodes that do not split are called Leaf or Terminal nodes.
<br>

**Pruning:** When we remove sub-nodes of a decision node, this process is called pruning. It is the opposite process of splitting.
<br>

**Branch/Sub-Tree:** A sub-section of an entire tree is called a branch or sub-tree.
<br>

**Parent and Child Node:** A node, which is divided into sub-nodes is called the parent node of sub-nodes where sub-nodes are the children of a parent node.

<br> 

#### <span style='color:Black'> **How does Decission Tree work?** </span>

##### **Decision Tree algorithm intuition**
1. For each attribute in the dataset, the Decision-Tree algorithm forms a node. The most important attribute is placed at the root node.
<br>

2. For evaluating the task in hand, we start at the root node and we work our way down the tree by following the corresponding node that meets our condition or decision.
<br>
3. This process continues until a leaf node is reached. It contains the prediction or the outcome of the Decision Tree.

<br>

##### **Criteria to split the data**
Decision tree uses 'entropy' or 'gini' selection criteria to split the data.


**Entropy**

Entropy measures the impurity in the given dataset.

<span style='color:blue'> $$Entropy = \sum_{i=1}^{C} - p_{i} \log_2 p_{i} $$ </span>

**Gini**

The Gini coefficient measures the inequality among values of a frequency distribution.
<br>
<span style='color:blue'> $$Gini = 1 - \sum_{i=1}^{C} (p_{i})^2$$ </span>

<br> 

#### <span style='color:Darkgreen'> **Advantages Of Decision Tree**</span>

<span style='color:Darkgreen'> - Simple to understand and to interpret. Trees can be visualized.</span>

<span style='color:Darkgreen'> - Requires little data preparation. Other techniques often require data normalization, dummy variables need to be created and blank values to be removed. Note however that this module does not support missing values.</span>

<span style='color:Darkgreen'> - Able to handle both numerical and categorical data and multi-output problems.</span>

<span style='color:Darkgreen'> - Uses a white box model. Results are easy to interpret.</span>

<span style='color:Darkgreen'> - Possible to validate a model using statistical tests. That makes it possible to account for the reliability of the model.</span>


<br> 

#### <span style='color:darkred'> **Disadvantages  Of Decision Tree**</span>

<span style='color:darkred'> - Decision-tree learners can create over-complex trees that do not generalize the data well. This is called overfitting. Mechanisms such as pruning, setting the minimum number of samples required at a leaf node or setting the maximum depth of the tree are necessary to avoid this problem. </span>

<span style='color:darkred'> - Decision trees can be unstable because small variations in the data might result in a completely different tree being generated. This problem is mitigated by using decision trees within an ensemble. </span>

<span style='color:darkred'> - Decision tree learners create biased trees if some classes dominate. It is therefore recommended to balance the dataset prior to fitting with the decision tree. </span>


<br>

#### <span style='color:Black'>**Decission Tree classifier in Python**</span>

##### <br> <span style = 'color:RebeccaPurple'>**Importing the libraries (modules)**</span>

```python
import pydotplus
from sklearn.tree import DecisionTreeClassifier 
from sklearn import datasets

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

##### <br> <span style = 'color:RebeccaPurple'>**Importing and exploring the data** </span>

```python
iris = datasets.load_iris() 
features = iris.data 
target = iris.target

print(features)
print(target)
```

<br> <span style = 'color:RebeccaPurple'>**Splitting the dataset into the Training set and Test set**</span>

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 0)
print(len(X_train))
print(len(X_test))
```

##### <br> <span style = 'color:RebeccaPurple'>**Building the Decision Tree model** </span>

```python
decisiontree = DecisionTreeClassifier(random_state=0) 
model = decisiontree.fit(X_train, y_train)
```

<br> <span style = 'color:RebeccaPurple'>**Predicting the Test set results**</span>

```python
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
```

<br> <span style = 'color:RebeccaPurple'>**Dot Data for the predictions**</span>

```python
from sklearn import tree

dot_data = tree.export_graphviz(decisiontree, out_file=None,
feature_names=iris.feature_names, 
class_names=iris.target_names
)
```

<br> <span style = 'color:RebeccaPurple'>**Drawing the Graph**</span>

```python
from IPython.display import Image

graph = pydotplus.graph_from_dot_data(dot_data) # Show graph

Image(graph.create_png()) 	
```

#### <span style='color:Black'> **Introduction** </span>

- Random forest is a supervised learning algorithm.
- It has two variations  one is used for classification problems and other is used for regression problems.
- It creates decision trees on the given data samples, gets prediction from each tree and selects the best solution by means of voting.

##### **Random Forest algorithm intuition**

<span style='color:darkgreen'>**Random forest algorithm intuition can be divided into two stages.**</span>

<span style='color:darkgreen'>


**First Stage:**
1. Randomly select $k$ features from a total of m features where $k < m$.
2. Among the $k$ features, calculate the node d using the best split point.
3. Split the node into child nodes using the best split.
4. Repeat 1 to 3 steps until l number of nodes has been reached.
5. Build forest by repeating steps 1 to 4 for n number of times to create $n$ number of trees.
</span>

<span style='color:darkgreen'>

**Second Stage:**
1. We take the test features and use the rules of each randomly created decision tree to predict the outcome and stores the predicted outcome.
2. Then, we calculate the votes for each predicted target.
3. Finally, we consider the high voted predicted target as the final prediction from the random forest algorithm.

</span>

<p align="center">
    <img src='./imgs/pic2.jpg' width = '70%'>
</p>

<span style='color:darkgreen'>

**Advantages Of Random Forest**
1. Random forest algorithm can be used to solve both classification and regression problems.
2. It is considered as very accurate and robust model.
3. Random forests takes the average of all the predictions made by the decision-trees, which cancels out the biases. So, it does not suffer from the overfitting problem.
4. Random forest classifier can handle the missing values. 
5. Random forest classifier can be used for feature selection.

</span>

<span style='color:red'>

**Disadvantages of Random Forest algorithm**

1. The biggest disadvantage of random forests is its computational complexity.
2. The model is difficult to interpret as compared to a decision-tree, where we can easily make a prediction as compared to a decision-tree.

</span>

#### <span style='color:Black'>**Random Forest in Python**</span>

<br> <span style = 'color:RebeccaPurple'>**Import libraries**</span>

```python
import pandas as pd
import numpy as np

# Loading the Penguins Dataset from Seaborn
import seaborn as sns

# Imputing missing numerical data
from sklearn.impute import SimpleImputer

from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

<br> <span style = 'color:RebeccaPurple'>**Importing the dataset**</span>

```python
df = sns.load_dataset('penguins')
print(df.head())
```

<p align="center">
    <img src='./imgs/pic3.jpg' width = '30%'>
</p>

```python
print(df.info())
```

<br> <span style = 'color:RebeccaPurple'>Dealing with Missing Data in Scikit-Learn</span>

```python
print(df.isnull().sum())
```

```python
# Create a SimpleImputer Class
imputer = SimpleImputer(missing_values=np.NaN, strategy='mean')

# Fit the columns to the object
columns = ['bill_depth_mm', 'bill_length_mm', 'flipper_length_mm', 'body_mass_g']
imputer=imputer.fit(df[columns])

# Transform the DataFrames column with the fitted data
df[columns]=imputer.transform(df[columns])
```

```python
# Dropping missing records in the sex column
df = df.dropna(subset=['sex'])
```

```python
print(df.isnull().sum())
```

<br> <span style = 'color:RebeccaPurple'>Dealing with Categorical Data in Scikit-Learn</span>

```python
# Mapping the sex variable to binary values
df['sex int'] = df['sex'].map({'Male': 0, 'Female': 1})
```

```python
# Checking unique values in the island feature
print(df['island'].unique())
```

<p align="center">
    <img src='./imgs/pic4.png' width = '40%'>
    <br>
    How one-hot encoding works in Pythons Scikit-Learn?
</p>

```python
# One-hot Encoding the Island Feature
one_hot = OneHotEncoder()
encoded = one_hot.fit_transform(df[['island']])
df[one_hot.categories_[0]] = encoded.toarray()
```

```python
# Dropping Unnecessary Columns
df = df.drop(columns=['island', 'sex'])
```

<br> <span style = 'color:RebeccaPurple'>**Splitting the dataset into the Training set and Test set**</span>

```python
# Splitting the data and creating a model
X = df.iloc[:, 1:]
y = df['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=100)
```

<br> <span style = 'color:RebeccaPurple'>**Create Random Forest Model**</span>

```python
forest = RandomForestClassifier(n_estimators=100, random_state=100)
# Fitting a model and making predictions
forest.fit(X_train,y_train)
predictions = forest.predict(X_test)
```

<br> <span style = 'color:RebeccaPurple'>**Evaluating the Performance of a Random Forest**</span>

```python
# Evaluating the model
print("Accuracy:",metrics.accuracy_score(y_test, predictions))
```

<br> <span style = 'color:RebeccaPurple'>**Visualizing Random Forest Decision Trees**</span>

```python
fig = plt.figure(figsize=(15, 10))
plot_tree(forest.estimators_[0], 
          feature_names=X.columns,
          class_names=df['species'].unique(), 
          filled=True, rounded=True)

plt.show()
```

```python
fig = plt.figure(figsize=(15, 10))
plot_tree(forest.estimators_[10], 
          feature_names=X.columns,
          class_names=df['species'].unique(), 
          filled=True, rounded=True)

plt.show()
```
